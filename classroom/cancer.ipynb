```python
y = df.diagnosis
```

```python
x = df.drop(columns=["id", "diagnosis", "Unnamed: 32"])
```

```python
y.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x,y)
```

```python
lr.score(x,y)
```

```python
lr.predict([x.iloc[0]])
```

```python
x.columns
```

```python
lr.coef_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(x.columns,lr.coef_[0])
```
